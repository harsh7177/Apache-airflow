```python
import os
import pandas as pd
col_specs=[(0,6),]
```

```python
names=['Row_id','Time_Stamp','Vehicle_Num','Toll_id','Toll_Code','Type_of_Payment','Vehicle_Code']
cname=['Row_id','Time_Stamp','Vehicle_Num','Vehicle_Type','No_of_axel','Vehicle_Code']
tname=['Row_id','Time_Stamp','Vehicle_Num','Veh_Type','No_of_axel','Toll_id','Toll_Code']
for file in os.listdir(os.getcwd()):
    if file.endswith('.txt'):
        txt_df=pd.read_fwf(os.path.join(os.getcwd(),file),
                           widths=[6,25,12,4,10,4,6],
                           header=None,index_col=0,
                           names=names)
    if file.endswith('.csv'):
        csv_df=pd.read_csv(os.path.join(os.getcwd(),file),index_col=0,names=cname)
                           
    if file.endswith('.tsv'):
        tsv_df=pd.read_csv(os.path.join(os.getcwd(),file),sep='\t',index_col=0,names=tname)
    
        

```

```python
d_df = pd.concat([txt_df, csv_df,tsv_df], ignore_index=True)
```
